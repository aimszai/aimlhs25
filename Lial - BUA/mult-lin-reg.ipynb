```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
y = np.array ([[93, 106, 146, 140, 158]]) .T
y
```

```python
x1 = np.array ([[2, 2, 4, 5, 7]]) .T
x1
```

```python
x2 = np.array ([[1.1, 1.9, 2, 1.5, 2.1]]) .T
x2
```

```python
X = np.hstack ((np.ones_like (x1), x1, x2))
X = np.hstack ((x1 - x1 + 1, x1, x2))
X
```

```python
beta_hat = la.inv (X.T @ X) @ X.T @ y

plt.figure(figsize=(10, 6))
plt.scatter(np.arange(1, y.shape[0] + 1), y, label='Actual y', color='blue', marker='o', s=100)
plt.plot(np.arange(1, y.shape[0] + 1), X @ beta_hat, label='Fitted y', color='red', linewidth=2)
plt.xlabel('Observation Index')
plt.ylabel('y')
plt.title(f'Linear Regression: y = {beta_hat[0,0]:.4f} + {beta_hat[1,0]:.4f}*x1 + {beta_hat[2,0]:.4f}*x2')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
